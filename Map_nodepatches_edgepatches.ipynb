**Mappings node to node-patch and edge to edge-patch**

For a given 2d triangulation (all triangles) computes:
- the mapping for each  vertix or node of the triangulation to the patches, all triangles sharing the vertex.
- the mapping for each edge of the triangulation to the pathces, all the trianles sharing that edge.

```python
from fenics import *
import csv
```

```python
# Cargar malla de prubea
mesh = UnitSquareMesh(2, 2)

# Cargar malla y_tube
# path_to_mesh= "/content/ytube_2d.h5"

# import dolfin
# mesh = dolfin.Mesh()
# hdf = dolfin.HDF5File(mesh.mpi_comm(), path_to_mesh, 'r')
# hdf.read(mesh, '/mesh', False)
# boundaries = dolfin.MeshFunction('size_t' , mesh, mesh.topology().dim() - 1)
# hdf.read(boundaries, '/boundaries')
# hdf.close()
```

```python
def save_to_file(file_path, data):
  # Save the data to the file
  with open(file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)
    print(f"Data has been saved to {file_path}")
  pass

def open_file(file_path):
  # Open and read the CSV file
  with open(file_path, 'r') as file:
    reader = csv.reader(file)
    data = [list(map(int, row)) for row in reader]
  return data
```

```python
Node2Elements = []

for v in vertices(mesh):
  # print("*** vertex id={} ***".format(v.index()))
  cells_v = []
  for c in cells(v):
    # print(c.index());
    cells_v.append(c.index())
  Node2Elements.append(cells_v)

Edge2Elements = []
for e in edges(mesh):
  # print("*** edge id={} ***".format(e.index()))
  cells_e = []
  for c in cells(e):
    # print(c.index())
    cells_e.append(c.index())
  Edge2Elements.append(cells_e)
```

```python
save_to_file(file_path='node2patch_unitsquare.csv', data=Node2Elements)

save_to_file(file_path='edge2patch_unitsquare.csv', data=Node2Elements)
```
